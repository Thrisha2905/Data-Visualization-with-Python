```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
X=pd.read_csv('canadian_immegration_data.csv')
X
```

```python
X.sort_values(['Total'], ascending = False, axis = 0, inplace = True)
X
```

```python
#generating area plots
years = list(map(str, range(1980, 2014)))
X_top5 = X.head()
X_top5 = X_top5[years].transpose()
Y=X_top5.head(40)
Y
```

```python
Y.plot(kind='area')
plt.title('Imigration trend of top 5 countries')
plt.ylabel('number of imigration')
plt.xlabel('years')
plt.show()
```

```python
#generating histogram
count, bin_edges = np.histogram(X['2013'])
X['2013'].plot(kind='hist',xticks = bin_edges)
plt.title('Histogram of Imigration from 195 countries in 2013')
plt.ylabel('Number of countries')
plt.xlabel('Number of Imigrants')
plt.show()
```

```python
#generating pie chart
X_continents = X.groupby('Continent', axis = 0).sum()
X_continents
```

```python
X_continents['Total'].plot(kind='pie')
plt.title('Immigrations to canada by continent [1980-2013]')
plt.show()
```

```python
import folium
#folium usage
#define the world map centered around canada with a low zoom level
world_map=folium.Map(location=[56.130,-106.35],zoom_start=4)
#display world map
world_map
```

```python
#marking on map
#generating map of canada
canada_map=folium.Map(location=[56.130,-106.35],zoom_start=4)
#add a red marker to Ontorio
#create a feature group
ontario=folium.map.FeatureGroup()
#style the feature group
ontario.add_child(
    folium.features.CircleMarker(
        [51.25,-85.32],radius = 5,color='red',fill_color='Red'
        )
    )
#add the feature group to the map
canada_map.add_child(ontario)
#label the marker
folium.Marker([51.25,-85.32],popup='Ontario').add_to(canada_map)
#display map
canada_map
```

```python
#generating choropleth maps
#create a plain world map
world_map=folium.Map(zoom_start=2,tiles='Mapbox Bright')
##geojaon file
world_geo=r'world_countries.jaon'
#generate choropleth map using the total
#population of each country to canada from 1980 to 2013
world_map.choropleth(
    geo_path=world_geo,
    data=X,
    columns=['Country','Total'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    legend_name='immigration to canada'
)
#display map
world_map
```
