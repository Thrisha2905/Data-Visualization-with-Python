```python
#we can also use the pandas function for plotting for the whole data set using the function kind=line
#lets take a sample dataset
data=pd.DataFrame({"Shoes":[75,60,85],"Price":[82,65,90]})
data
```

```python
data.plot(kind='line')
```

```python
plt.xlabel='X'
plt.ylabel='Y'
plt.title('Shoes prices')
```

```python
data.plot.hist()
```

```python
data.plot.scatter(x='Shoes',y='Price')
```
